```python
# Imports
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
```

```python
# Test datasets
def get_dataset(lang):
    """
    Returns scaled dataset for English or French. 
    Scaling is done by dividing by the total number of characters in respective book.
    :param lang: language
    :returns: Scaled dataset, scaling factor
    """
    if lang.lower() == 'eng' or lang.lower() == 'english':
        DS = [[35680, 2217],
              [42514, 2761],
              [15162, 990],
              [35298, 2274],
              [29800, 1865],
              [40255, 2606],
              [74532, 4805],
              [37464, 2396],
              [31030, 1993],
              [24843, 1627],
              [36172, 2375],
              [39552, 2560],
              [72545, 4597],
              [75352, 4871],
              [18031, 1119]
             ]
    else:
        DS = [[36961, 2503],
              [43621, 2992],
              [15694, 1042],
              [36231, 2487],
              [29945, 2014],
              [40588, 2805],
              [75255, 5062],
              [37709, 2643],
              [30899, 2126],
              [25486, 1784],
              [37497, 2641],
              [40398, 2766],
              [74105, 5047],
              [76725, 5312],
              [18317, 1215]
             ]
        
    # Scale the dataset
    scale = max(np.sum(DS, axis=0))
    DS = np.divide(DS, scale)
    DS = np.asarray(DS)
    
    return DS, scale
```

```python
def linreg(X, y, w, version='SGD', learning_rate=1, epsilon=1e-3, maxit=10000):
    """
    Method for finding parameters for linear regression, terminates if gradiant of the loss evaluated with 
    current weights w.
    :param DS: Dataset DS
    :param w: initial weights w
    :(optional) param version: SGD or batch
    :(optional) param learning_rate: step length
    :(optional) param epsilon: termination crit
    :(optional) param maxit: maximum number of iterations
    :return w: optimal weights w
    """
    num_points, _ = np.shape(DS)
    if num_points == 0:
        raise Exeption('Not enough points. number of points was 0')
    
    if num_points == 1:
        print('Only one point, the intercept is set to 0 the slope is calculated thereafter.')
        return 0, DS[1]/DS[0]
    
    else:
        converged = False
        update_order = np.arange(num_points)
        for ep in range(maxit):
            if version == 'SGD':
                rnd.shuffle(update_order)
                for update in update_order:
                    grad_loss = (y[update] - X[update] @ w)*X[update].T
                    w += learning_rate * grad_loss 
                if np.linalg.norm(X @ w) < epsilon:
                    print('Terminated at episode:{}'.format(ep+1))
                    converged = True
                    break
            else:
                
                grad_loss = X.T @ (y - X @ w)
                w += learning_rate * grad_loss/num_points
                
                if np.linalg.norm(X @ w) < epsilon:
                    print('Terminated at episode:{}'.format(ep+1))
                    converged = True
                    break
        if not converged:
            print('Termination criterion not meet, code terminated due to too many iterations. Values returned may not be optimal.')
        return np.asarray(w) 
        
```

```python
# Run and plot linear regression.
DS, scale = get_dataset('eng')
y = DS[:, -1]
observations = len(y)
X = np.concatenate((np.ones((observations, 1)), DS[:, :-1]), axis=1)
w = linreg(X, y, [0, 0], version='SGD')

x = DS[:, 0]
y = DS[:, 1]
plot_x = x*scale
plot_y = y*scale

plt.figure(figsize=[12.8, 9.6])

plt.scatter(plot_x, plot_y)

x_min_plot = min(plot_x)
x_max_plot = max(plot_x)
xx_plot = [x_min_plot, x_max_plot]

yy_plot = [np.array([1, x_min_plot]) @ w, np.array([1, x_max_plot]) @ w]
print(w)

plt.plot(xx_plot, yy_plot)
plt.xlabel('Total words', fontsize=16)
plt.ylabel('Number of a\'s', fontsize=16)
plt.show()
```
